# Príprava dát

```python
# nahrádenie kategorických hodnôt číselnými hodnotami 

student['gender'].replace('M', 0,inplace=True) # nahradenie hodnôt z triedy gender hodnotu M na 0
student['gender'].replace('F', 1,inplace=True) # nahradenie hodnôt z triedy gender hodnotu F na 1

# použijeme get_dummies na premenenie kategorických hodnôt a neskôr ich budeme zreťazovať, a  priraďovať do potrebnej postupnosti
nat = pd.get_dummies(student['NationalITy']) # nastavenie premennej NationalIty na nat
sid = pd.get_dummies(student['StageID']) # nastavenie premennej StageID na sid
gid = pd.get_dummies(student['GradeID']) # nastavenie premennej GradeID na gid
secid = pd.get_dummies(student['SectionID']) # nastavenie premennej SectionID na secid
topic = pd.get_dummies(student['Topic']) # nastavenie premennej Topic na topic
semester = pd.get_dummies(student['Semester']) # nastavenie premennej Semester na semester
rel = pd.get_dummies(student['Relation']) # nastavenie premennej Relation na rel
pas = pd.get_dummies(student['ParentAnsweringSurvey']) # nastavenie premennej ParentAnsweringSurvey na pas
pss = pd.get_dummies(student['ParentschoolSatisfaction']) # nastavenie premennej ParentschoolSatisfaction na pss
sab = pd.get_dummies(student['StudentAbsenceDays']) # nastavenie premennej StudentAbsenceDays na sab
```

```python
# odstránenie stĺpcov, ktoré sa nahradili o krok vyššie
student.drop(['NationalITy','PlaceofBirth','PlaceofBirth','StageID','GradeID','SectionID','Topic','Semester',
              'Relation','ParentAnsweringSurvey','ParentschoolSatisfaction','StudentAbsenceDays'],axis=1,inplace=True)
```

```python
# zretazenie hodnôt do študent dátoveho rámcu
student = pd.concat([student,nat,sid,gid,secid,topic,semester,rel,pas,pss,sab],axis=1)
```

```python
student.columns # zobrazenie stĺpcov
```

```python
Label = student['Class'] # nastavenie Label hodnotou Class zo študent dátového rámcu
Features = student[['gender', 'raisedhands', 'VisITedResources', 'AnnouncementsView',
       'Discussion','Father', 'Mum', 'No', 'Yes', 'Bad', 'Good',
       'Above-7', 'Under-7']] # nastavenie parametrov 
```

```python
# základné funkcie na odstránenie priemeru a nastavenie jednotlivých škálovacích rozptylov
scaler = StandardScaler()  # nastavenie funkcie StandarScaler (funkcia škálovania) pre hodnotu scaler
scaler.fit(Features) # nastavenie škálovacej hodnoty pre triedu Features
scaled = scaler.transform(Features) # zmenenie hodnôt výpisu
```

```python
# rozdelenie dát na 20% testovacích a 80% trenovacích dát s náhodným stavom 42
X = scaled #nastavenie hodnoty X
y = Label # nastavenie hodnoty Y
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42) #priradenie hodnôt pre testovacie a trenovacie dát 
```

```python
err = [] # pole, kde sa ukladajú všetky chyby

for i in range(1,20): # cyklus skúša všetky chyby 
    rfe = RandomForestClassifier(n_estimators=i*10,random_state=42) # vytvorí nám RFC model s počtom odhadov i*10 
    rfe.fit(X_train,y_train) # dáta sa prispôsobia modelu 
    errpred = rfe.predict(X_test) # predpovedá hodnotu
    err.append(np.mean(errpred != y_test)) # pridá hodnotu do poľa 
    
    
# vykreslí hodnoty chybovosti pomocou metódy odhadu, ktorá je vytvorená vyššie. 
plt.figure(figsize=(15,7)) # nastavenie veľkosti

plt.plot(range(1,20),err,color='blue',linestyle='dotted',marker='o',markerfacecolor='red',markersize=10)#vykreslenie hodnôt
plt.title = ('Počet odhadov VS chybovosť')  # počet odhadov vs chybovosť
plt.xlabel =  'Odhady'  #X label
plt.ylabel= 'chybovosť' # Y label
plt.show()
```

# RandomForestClassifier

```python
# model RFC 
rfc = RandomForestClassifier(n_estimators=80,max_features='auto', max_depth=9,min_samples_leaf=1,# nastavenie limitov pre model RFC, 
                             min_samples_split=2,bootstrap=True, random_state = 42)
rfc.fit(X_train,y_train) # nastavenie hodnoty pre model
rfcpred = rfc.predict(X_test) # nastavenie hodnoty pre model
```

```python
print('Random Forest Classifier' + '\n') # vykreslenie nastroja RFC

print(classification_report(y_test,rfcpred))# vykreslenie hodnoty 

print('\n')# vytvorenie odseku

print('Confusion matrix') # vykreslenie matice kde vidieť, v akých bodoch fungoval model najepšie

ax = sns.heatmap(confusion_matrix(y_test,rfcpred),cmap='YlGn_r',annot=True,fmt='g'  ) #  vykreslenie matice s hodnotami po boku
ax.set_title ('matica'); #  nazov matice
ax.set_ylabel ('values'); #  y strana matice
ax.set_xlabel(' predicted values') # x strana matice
```

# k-Nearest Neighbors algoritmus

```python
 # k-Nearest Neighbors algoritmus (KNN)
err_rate = [] # pole, kde sa ukladajú všetky chyby

for i in range(1,40): # cyklus skúša všetky chyby 
    knn = KNeighborsClassifier(n_neighbors=i) # vytvorí objekt KNN s hodnotou i
    knn.fit(X_train,y_train) # nastavýmeho na model
    pred_i = knn.predict(X_test) # predpovedana hodnota 
    err_rate.append(np.mean(pred_i != y_test)) #pridá hodnotu do pola 
    
# vykreslí hodnoty chybovosti pomocou metódy odhadu, ktorá je vytvorená vyššie. 
plt.figure(figsize=(20,10)) # nastavenie veľkosti 
plt.plot(range(1,40),err_rate,color='blue',linestyle='dotted',marker='o',markerfacecolor='red',markersize=8)#vykreslenie hodnoty
plt.title = 'K Values VS Error Rates' #nazou 
plt.xlabel = 'K Value' #x label
plt.ylabel= 'Error Rate' # y label
plt.show()
```

```python
# modelKNN
knn = KNeighborsClassifier(n_neighbors=21,p=10) #  nastavenie parametrov 
knn.fit(X_train,y_train) # trenovacie data 
knnpred = knn.predict(X_test) #  testovacie data 
```

```python
print('K Nearest Neighbours' + '\n') # vypísanie hodnot KNN
print(classification_report(y_test,knnpred))  # vykreslí hodnoty

print('\n')# správy odsek

print('Confusion matrix') # vykreslenie matice 
ax = sns.heatmap(confusion_matrix(y_test,knnpred),cmap='Oranges_r',annot=True,fmt='g') # vykreslí maticu s hodnotami po boku
ax.set_title ('matica');#  nazov matice
ax.set_ylabel ('values'); #  y strana matice
ax.set_xlabel(' predicted values')  # x strana matice
```

# k-fold cross-validation

```python
#vybrate modeli
models = [] # vytvorenie pražnej množiny
models.append(('LR', LogisticRegression(solver='liblinear', multi_class='ovr'))) # vloženie LR do premenej models
models.append(('RFC', RandomForestClassifier()))# vloženie RFC do premenej models
models.append(('KNN', KNeighborsClassifier()))# vloženie KNN do premenej models
models.append(('CART', DecisionTreeClassifier()))# vloženie CART do premenej models
models.append(('SVM', SVC(gamma = 'auto')))# vloženie SVM do premenej models

results = []
names = []
for name, model in models:
    kfold = StratifiedKFold(n_splits = 10, random_state=1, shuffle = True)
    cv_results = cross_val_score(model, X_train, y_train, cv = kfold, scoring='accuracy')
    results.append(cv_results)
    names.append(name)
    print('%s: %f (%f)' % (name, cv_results.mean(), cv_results.std()))
```
